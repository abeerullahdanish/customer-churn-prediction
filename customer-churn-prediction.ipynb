```python
# Ye function 'SeniorCitizen' column mein 1 ko "Yes" aur 0 ko "No" mein convert karta hai taake data zyada readable ho jaye.

def conv(value):
  if value ==1:
    return "Yes"
  else:
    return "No"
df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
# Ye count plot dikhata hai ke kitne customers ne service chhor di aur kitne nahi, aur har bar ke upar exact count label bhi show karta hai.
plt.figure(figsize=(6, 6))  # Figure size set karo pehle
# Countplot with custom colors
ax = sns.countplot(x='Churn', data=df, palette='Set2')  # Set2 ek acha pastel color palette hai
# Bar labels har bar ke upar
ax.bar_label(ax.containers[0])
# Title add keyia howa ha
plt.title('Count of Customers by Churn', fontsize=16)
plt.show()
```

```python
# 'Churn' ki values gin kar, chhota figure size (4x4 inch) set keyia howa ha , phir pie chart banayia ha  jisme percentage aur title ho. Pie ko gol banaya gaya hai.
churn_counts = df['Churn'].value_counts()
# Figure size set karo
plt.figure(figsize=(4, 4))
# Pie chart banao
plt.pie(churn_counts, labels=churn_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Customer Churn Percentage Distribution')
plt.axis('equal')  # Pie ko gol banata hai
plt.show()
```

```python
# Gender counts  bhe nikalain
gender_counts = df['gender'].value_counts()
plt.figure(figsize=(4,4))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['#4daf4a', '#377eb8'])
plt.title('Gender Percentage Distribution', fontsize=16)
plt.axis('equal')  # Circle banane ke liye
plt.show()
```

```python

# 1. Crosstab for counts of SeniorCitizen vs Churn
cross_tab = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# 2. Plot stacked bar chart
ax = cross_tab.plot(kind='bar', stacked=True, figsize=(8,5), colormap='Pastel1')

# 3. Add percentages and labels
total_per_bar = cross_tab.sum(axis=1)

for i, (idx, row) in enumerate(cross_tab.iterrows()):
    cumulative = 0
    for j, val in enumerate(row):
        height = val
        percent = val / total_per_bar[idx] * 100
        ax.text(i, cumulative + height / 2, f'{val}\n({percent:.1f}%)', ha='center', va='center', fontsize=10)
        cumulative += height

# 4. Titles and labels
plt.title('Senior Citizen vs Churn (Stacked Bar Chart with Percentages)')
plt.xlabel('Senior Citizen (0=No, 1=Yes)')
plt.ylabel('Count')
plt.legend(title='Churn')
plt.xticks(rotation=0)
plt.show()
```

```python
# Gender aur Churn ke darmiyan talluq ko dikhane ke liye crosstab banayia ha
# Phir stacked bar chart plot banayia ha  jisme har gender ke liye churn ke Yes/No counts stacked ha
#  bar segment ke upar uska count aur percentage label bhi show keyia ha  taake data asani se samajh aaye
# Chart ki title, axis labels, ticks aur legend bhi set kiye howay  hain
gender_churn = pd.crosstab(df['gender'], df['Churn'])

# Stacked bar chart plot karo
ax = gender_churn.plot(kind='bar', stacked=True, figsize=(7,5), colormap='Pastel1')

# Percentages aur labels add karo
totals = gender_churn.sum(axis=1)
for i, (idx, row) in enumerate(gender_churn.iterrows()):
    cumulative = 0
    for j, val in enumerate(row):
        percent = val / totals[idx] * 100
        ax.text(i, cumulative + val / 2, f'{val}\n({percent:.1f}%)', ha='center', va='center')
        cumulative += val

plt.title('Gender vs Churn (Stacked Bar Chart with Percentages)')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Churn')
plt.show()
```

```python
# Boxplot banaya ha hai jo 'MonthlyCharges' ke distribution ko churn karne walon (Yes) aur na karne walon (No) ke liye compare karta hai.
# 'Set3' palette use kiya  hai taake colors clearly dikhain.
# Title aur axis labels bhi add kiye howey hain.
plt.figure(figsize=(8,5))
sns.boxplot(x='Churn', y='MonthlyCharges', data=df, palette='Set3')
plt.title('Monthly Charges Distribution by Churn')
plt.xlabel('Churn')
plt.ylabel('Monthly Charges')
plt.show()
```

```python
# 'Contract' ke mukhtalif types ka countplot banaya hai, jisme 'Churn' (Yes/No) ke hisaab se rang alag hain.
# Har bar ke upar uski counting labels bhi dikhayi gayi hain.
# Title aur axis labels set kiye howey  hain.
plt.figure(figsize=(8,5))
ax = sns.countplot(x='Contract', hue='Churn', data=df, palette='coolwarm')
# Bar labels
for container in ax.containers:
    ax.bar_label(container)
plt.title('Contract Type vs Churn')
plt.xlabel('Contract Type')
plt.ylabel('Count')
plt.show()
```

```python
# Har Payment Method ke liye churn ('Yes' aur 'No') ki counting ke liye cross-tabulation banai ha .
# Har payment method ka total churn count add kiya ha  aur total ke hisaab se sort kiya.
# Sirf 'Yes' yani churn karne walon ki tadaad nikal ke pie chart banaya ha.
# Pie chart mein payment method ke mutabiq churn percentage dikhayi gayi hai.
payment_churn = pd.crosstab(df['PaymentMethod'], df['Churn'])
payment_churn['Total'] = payment_churn.sum(axis=1)
payment_churn = payment_churn.sort_values(by='Total', ascending=False)

# Pie chart for churn=yes counts by PaymentMethod
churn_yes = payment_churn['Yes']

plt.figure(figsize=(4,4))
plt.pie(churn_yes, labels=churn_yes.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Churn Distribution by Payment Method')
plt.axis('equal')
plt.show()
```

```python
# 'OnlineSecurity' ke categories ka countplot banaya gaya hai, jisme 'Churn' (Yes/No) ke hisaab se colors alag hain.
# Har bar par count labels bhi banayia  gaye hain. Title aur axis labels bhi diye gaye hain.
plt.figure(figsize=(8,5))
ax = sns.countplot(x='OnlineSecurity', hue='Churn', data=df, palette='Set1')

for container in ax.containers:
    ax.bar_label(container)

plt.title('Online Security Subscription vs Churn')
plt.xlabel('Online Security')
plt.ylabel('Count')
plt.show()
```
